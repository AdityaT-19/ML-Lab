# Data Visualization

**Packages Import**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

**Data Import**

```python
data = pd.read_csv("xyz_data.csv")
X = data['X']
Y = data['Y']
Z = data['Z']
```

**Scatter Plot**

```python
plt.scatter(X,Y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot')
```

**Box Plot**

```python
sns.boxplot(data=data)
plt.title('Box Plot')
```

**Heat Map**

```python
sns.heatmap(data.corr(),annot=True,cmap='coolwarm')
plt.title('Heat Map')
```

**Contour Plot**

```python
plt.tricontour(X,Y,Z,cmap='jet')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Contour Plot')
```

**3D Surface Plot**

```python
ax = plt.axes(projection="3d")
ax.plot_trisurf(X,Y,Z,cmap='jet_r')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('3D Surface Plot')
```
